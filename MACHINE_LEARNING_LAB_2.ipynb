Objective-5 :
To study relationships between variables using basic exploration.
Tasks:
1. Compute correlation matrix.
2. Identify highly correlated variables.
3. Plot scatter plot between age and maximum heart rate.
4. Interpret relationships.

```python
correlation_matrix = df.corr(numeric_only=True)
display(correlation_matrix)
```

```python
import numpy as np

high_corr_threshold = 0.7

highly_correlated_pairs = []

# Get unique pairs from the upper triangle of the correlation matrix
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > high_corr_threshold:
            col1 = correlation_matrix.columns[i]
            col2 = correlation_matrix.columns[j]
            corr_value = correlation_matrix.iloc[i, j]
            highly_correlated_pairs.append((col1, col2, corr_value))

if highly_correlated_pairs:
    print(f"Highly correlated variables (absolute correlation > {high_corr_threshold}):")
    for pair in highly_correlated_pairs:
        print(f"- {pair[0]} and {pair[1]}: {pair[2]:.2f}")
else:
    print(f"No highly correlated variable pairs found with an absolute correlation greater than {high_corr_threshold}.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.scatter(df['Age'], df['Max HR'], alpha=0.6)
plt.title('Scatter Plot of Age vs. Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (Max HR)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Correlation Matrix:

The correlation matrix did not reveal any highly correlated variable pairs (absolute correlation greater than 0.7) among the numerical features. This suggests that there are no extremely strong linear relationships between most of the numerical variables in this dataset. However, there might be moderate correlations or non-linear relationships that are not captured by this simple linear correlation analysis.
Age vs. Maximum Heart Rate Scatter Plot:

The scatter plot of 'Age' vs. 'Maximum Heart Rate' (Max HR) likely shows a general trend where Maximum Heart Rate tends to decrease as Age increases. This is a common physiological observation: younger individuals typically have higher maximum heart rates than older individuals. While there might be some spread in the data, the overall pattern should reflect this inverse relationship.
